```python
# 
import pandas as pd
pd.set_option('display.max_rows', 100)
df_nute_pro = pd.read_csv(r'../data/식재료별영양소함량.csv',encoding='utf_8')
df_nute_pro.drop('1회 섭취 참고량',axis=1,inplace=True)
df_nute_pro.drop('식품중량',axis=1,inplace=True)
df_nute_pro
```

```python
df_nute_pro.columns=['식재료명','에너지(kcal)','탄수화물(g)','단백질(g)','지방(g)']
```

```python
columns_to_select = ['식품명','에너지(kcal)','탄수화물(g)','단백질(g)','지방(g)']
df_nute_farm = pd.read_csv(r'../data/농산물.csv',encoding='utf_8',usecols=columns_to_select)
df_nute_aqua = pd.read_csv(r'../data/수산물.csv',encoding='utf_8',usecols=columns_to_select)

df_nute_aqua.rename(columns={'식품명':'식재료명'},inplace=True)
df_nute_farm.rename(columns={'식품명':'식재료명'},inplace=True)
```

```python
cols = ['식재료명']+['에너지(kcal)']+['탄수화물(g)']+['단백질(g)']+['지방(g)']
df_nute_aqua=df_nute_aqua[cols]
df_nute_pro=df_nute_pro[cols]
df_nute_farm =df_nute_farm[cols]
display(df_nute_aqua)
display(df_nute_pro)
display(df_nute_farm)
```

```python
import numpy as np
df_nute_concat= pd.concat([df_nute_aqua,df_nute_farm,df_nute_pro],ignore_index=True)
df_nute_concat
```

```python
df_nute_concat.to_csv(r'../data/식재료별영양소함량.csv',index=False,encoding='utf-8')
```
